Scraping the Blog

```python
import bs4
import requests 
import re 
import json
```

```python
url = "https://www.moneycontrol.com/news/business/earnings/reliance-standalone-december-2023-net-sales-at-rs-127695-00-crore-up-1-36-y-o-y-12090761.html"
request = requests.get(url)
soup = bs4.BeautifulSoup(request.text,'html.parser')
```

```python
all_script = soup.find_all('script',attrs={'type':'application/ld+json'})
type(all_script)
```

```python
len(all_script)
```

```python
print(all_script[2])
```

```python
raw_article_str = all_script[2].get_text().replace('\r\n',' ') #r:raw str,n: new line
print(raw_article_str)
```

```python
parts = re.split(r"""("[^"]*"|'[^']*')""", raw_article_str)
parts 
```

```python
parts[::2] = map(lambda s:"".join(s.split()),parts[::2])
print(parts)
```

```python
article_str = "".join(parts)
article_str
```

```python
article_str = article_str[1:]
article_str = article_str[:-1]
print(article_str)
```

```python
# Identify the position of the problematic character
problematic_index = 3733

# Remove the problematic character from the JSON string
fixed_article_str = article_str[:problematic_index] + article_str[problematic_index + 1:]

# Attempt to parse the fixed JSON string
try:
    article_dict = json.loads(fixed_article_str)
    print(article_dict)
except json.JSONDecodeError as e:
    print("JSON decoding error:", e)
```

```python
all_tags = soup.findAll('div',attrs={'class':'tags_first_line'})
print(all_tags)
```

```python
lst_all_tags = []
for i in all_tags:
    lst_all_tags.append(i.get_text())
print(lst_all_tags)
```

```python
tags = lst_all_tags[0].replace('Tags:','')
tags = tags.replace('\n','')
tags
```

```python
tags = tags.split('#')
tags 
```

```python
tags = tags[1:]
print(tags)
```

```python
tags = ', '.join([str(elem).strip() for elem in tags])
print(tags)
```

```python
article_dict['tags'] = tags
print(article_dict)
```

```python
with open('data.json', 'w') as fp:
    json.dump(article_dict, fp)
```
